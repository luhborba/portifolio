```python
banco.rename(columns={
    'Data de Adessão' : 'Data de Adesão'
},inplace=True)
```

```python
banco.head(1)
```

```python
#Convertendo campo data
banco['Data de Adesão'] = pd.to_datetime(banco['Data de Adesão'])
```

```python
banco['Data de Adesão'].dtypes
```

```python
#Extraindo Ano e Mes
banco['Mes'] = pd.DatetimeIndex(banco['Data de Adesão']).month
banco['Ano'] = pd.DatetimeIndex(banco['Data de Adesão']).year

banco.head()
```

```python
#Tabela Analitica
Analise_Agrupada = banco.groupby(by=['Pais','Ano','Mes','Empresa']).count()['ID'].reset_index()
Analise_Agrupada.head()
```

```python
#Filtrando a tabela atraves do Loc
Analise_Brazil = Analise_Agrupada.loc[
    Analise_Agrupada['Pais'] == 'Brazil'
]
Analise_Brazil
```

```python
banco['Valor $'] = pd.to_numeric(banco['Valor $'].apply( lambda Linha: Linha.replace('$','')))
```

```python
banco.head()
```

```python
#Tabela Analitica
Analise_Pais = banco.groupby(by=['Pais']).sum()['Valor $'].reset_index()
Analise_Pais.head()
```

```python
Analise_Valor = Analise_Pais.sort_values('Valor $',ascending=False)
plt.figure(figsize=(15,6))
plt.title('Analise de Valor por Pais Top 8')
plt.plot(Analise_Valor['Pais'].head(8),Analise_Valor['Valor $'].head(8));
plt.xticks(rotation=45,ha='right');
plt.ylabel('Valor $ (Bilhões)')
```
